```python
# we are importing some of the important libraries for our analuysis as below:

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

# We are storing the data in batter data in dictionary format
Batter_data={"Batter_1_score":[56,19,0,68,16,67,85,101,78],
             "Batter_2_score":[60,58,38,71,69,66,68,41,67],
             "Temprature"    :[31,33,30,35,34,15,10,7,11] }
    
    
Batter_data
```

```python
# Here we are creating dataframe of same Data using pandas library as below:
Batter_df = pd.DataFrame(Batter_data)
Batter_df
```

```python
#We are using Describe function here to get all the central tendacy parameters information like mean standard deviation and all
Batter_df.describe()
```

### we can see from above table that batter_2 has average score of 59.77 as comapre to 54.44 of Batter_1 and standard deviation of 12.245 as comapre to 34.81 of Batter_1..it means score is less spreaded in case of Batter_2 and looks more stable than Batter_1...also Batter_1 is out on zero once..and Batter_2 minimum score is 38 which is quite good.

```python
# We are plotting grouped bar plot to see comaprison between both Batters_scores in different temperature range
Batter_df.plot(kind="bar",y=["Batter_1_score","Batter_2_score"],x="Temprature")
plt.show()
```

```python
plt.boxplot([Batter_df["Batter_1_score"],Batter_df["Batter_2_score"]])
plt.xlabel('Batter')
plt.ylabel('Scores')
plt.title('Box plot for two cricketers')
plt.show()

print("--------we can also notice batter_1 score is varying too much while batter_2_score is stable around  50 to 70 score\n")

print("--------We can see some outliers in Batter_2_score boxplot around 40 score and 50 percentile is around same of both that is 65 to 70")
```

# Lets See if there is any Linear relationship between Batters Scores and Temperature

```python

# here we are sorting Temperature column by ascending order so we can get the idea of data
batter_sort =Batter_winter.sort_values(by=['Temprature'])
batter_sort
```

```python
# We are plotting grouped bar plot to see comaprison between both Batters_scores in low temperature range
batter_sort.plot(kind="bar",y=["Batter_1_score","Batter_2_score"],x="Temprature")# plotting bar plot between batters_scores and tempt.

# we are plotting line graph for 2 batters _scores seperately with temperature to see any relationshipas below;
batter_sort.plot(x = 'Batter_1_score', y = 'Temprature')
batter_sort.plot(x = 'Batter_2_score', y = 'Temprature')
plt.show()
```

##  By above graphs we are getting some negative linear relationship between Batter_1_score and as Batter_1 score is increasing Linearly as Temperature Decreasing while Batter_2  score is decresing and that too not that much linear way..Lets see Correlation values of both batters respective to temperature using corr() function.

```python
# We are looking for correlation value between these variables using corr() function.

batter_sort.corr()
```

## so we can see the correlation value between Batter_1_score and temperature is -0.97 which is almost near to -1, which obeys negative liner relationship as shown in above graph
